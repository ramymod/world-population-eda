```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = df.drop(columns=[ 'World Population Percentage' , 'Rank'])
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[['1970 Population' , '1980 Population' , '1990 Population', '2000 Population', '2010 Population' , '2015 Population' , '2020 Population' , '2022 Population']].sum()
```

```python
top_10_70 = df.sort_values(by='1970 Population' , ascending=False).head(10)
```

```python
top_10_22 = df.sort_values(by='2022 Population' , ascending=False).head(10)
```

```python
sns.barplot(data=top_10_70, x='Country/Territory', y='1970 Population', palette='viridis')
plt.title('Top 10 Most Populous Countries in 1970')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=top_10_22, x='Country/Territory', y='2022 Population', palette='viridis')
plt.title('Top 10 Most Populous Countries in 1970')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
continent_pop = df.groupby('Continent')['2022 Population'].sum()

plt.figure(figsize=(8,8))
plt.pie(continent_pop, labels=continent_pop.index, autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution by Continent')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Area (km²)', y='Density (per km²)', hue='Continent')
plt.title('Area vs Population Density')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```
